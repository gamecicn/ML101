```python
import torch 
import numpy as np
```

# Empty Tensor

```python
# Create a 3D tensor, without initialize

x = torch.empty(1, 2, 3) # the left value is the outer size
print(x)
```

# One Tensor

init all value to 1 

```python
x = torch.ones(2, 3, dtype=torch.float16) # the left value is the outer size
print(x)
print(x.size())
```

# Creaet a concret tensor

```python
x = torch.tensor([2.5, 0.1])
print(x.size())
```

# Tensor operations

```python
x = torch.rand(2, 2)
y = torch.rand(2, 2)
print(x)
print(y)

z = x + y
print(z)
```

```python
print(torch.add(x, y))
```

In pytorch fucntion end with _ means inplace opertion

```python
print(y.add_(x))
```

```python
y = torch.rand(2, 2)
print(torch.mul(x, y))
```

# Slice Tensor

```python
x = torch.rand(5, 3)
print(x)

print(x[:, 0])
```

# Reshape

```python
x = torch.rand(4, 4)
print(x)

y = x.view(16)
print(y)

z = x.view(2, 8)
print(z)

q = x.view(2, -1)
print(q)
```

# Numpy and Tensor

## Tensor to numpy

```python
a = torch.ones(5)
print(a)

b = a.numpy()
print(b)
```

**Be Careful  !!! In this way, the Numpy and Tensor using the same memory,change one value will also change the other one.**

```python
a.add_(1) # remember funtion with _ means inplace operation
print(a)
print(b)
```

# Numpy to Tensor

```python
a = np.ones(5)
print(a)

b = torch.from_numpy(a)
print(b)

a += 1
print(a)
print(b)
```

# You cannot conver a GPU tensor to numpy

```python
if torch.cuda.is_available():
  device = torch.device("cuda")
  x = torch.ones(5, device=device)
  y = torch.ones(5)
  y = y.to(device)
  z = x + y

  print(z)

  # If you want to convert a GPU tensor, send it to CPU at first
  print(z.to("cpu").numpy())

  print(z.numpy())
```
